```python
intClass=[]
floatClass=[]
objectClass=[]
allClass=AllData.columns
def classifyType(allCol):
    for col in allCol:
        thisClass=AllData[col].dtypes
        if(thisClass=='int64'):
            intClass.append(col)
        elif(thisClass=='float64'):
            floatClass.append(col)
        elif(thisClass=='object'):
            objectClass.append(col)
        else:
            raise Exception('Un-handdle data type')
```

```python
classifyType(allClass)
```

```python
AllData[intClass].info()
```

```python
AllData[floatClass].info()
```

```python
AllData[floatClass]=AllData[floatClass].fillna(AllData.mean())
```

```python
AllData[objectClass].info()
```

```python
AllData[objectClass]=AllData[objectClass].fillna('None')
```
